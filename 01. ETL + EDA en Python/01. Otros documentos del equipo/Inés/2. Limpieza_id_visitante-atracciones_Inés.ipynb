# Limpieza id_visitante_atracciones.json




## 1 - Importación y lectura del df

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json('/content/id_visitante-atracciones.json')
```

```python
#Como no puedo abrirlo con pandas, abro el archivo de esta forma para poder verlo.
import json
with open('id_visitante-atracciones.json', 'r', encoding="utf-8") as file:
    data = json.load(file)
```

```python
#Confirmamos que lo que tenemos es un diccionario
type(data)
```

```python
#Para convertirlo en dataframe, vamos a hacer un bucle. Hacemos un primer bucle para las claves, y luego un segundo para las listas dentro del valor del diccionario.
datos_procedencia = []

for id , lista in data.items():
  for elemento_listas in lista:
     elemento_listas["id"] = int(id)
     datos_procedencia.append(elemento_listas)
```

```python
df_procedencia=pd.DataFrame(datos_procedencia)
df_procedencia.head()
```

## 2 - Limpieza del df

```python
#Vemos la info del df
df_procedencia.info()
```

```python
df_procedencia.shape
```

### Comprobación de nulos

```python
#Comprobamos si hay nulos
df_procedencia.isnull().sum()
```

```python
#Vamos a ver qué porcentaje de nulos hay en el df
print(f"Porcentaje de nulos en el df: {round((1138+96)/35887 * 100, 1)}%")
```

```python
#Hago una copia del df con nulos
df_nulos = df_procedencia
df_nulos.head()
```

```python
#Al ser solo un 3.4% de la información, decidimos eliminar estas filas con nulos.
df_procedencia = df_procedencia[df_procedencia['atraccion'].notna()]
df_procedencia = df_procedencia[df_procedencia['comienzo_atraccion'].notna()]
df_procedencia.isnull().sum()
```

```python
df_procedencia.head()
```

### Ordenar el df y confirmar que los valores son del tipo correcto - df sin nulos

```python
#Vamos a cambiar la columna 'comienzo atraccion' por enteros
df_procedencia['comienzo_atraccion'] = df_procedencia['comienzo_atraccion'].astype(int)
df_procedencia.info()
```

```python
df_procedencia.head()
```

```python
#Cambiamos el nombre 'id por 'id_visitante', que es lo que corresponde
df_procedencia.rename(columns={'id': 'id_visitante'}, inplace=True)
df_procedencia.head()
```

### Ordenar el df y confirmar que los valores son del tipo correcto - con nulos

```python
#Vamos a cambiar la columna 'comienzo atraccion' por enteros, ya que todos son .0 y no nos beneficia
df_nulos['comienzo_atraccion'] = pd.to_numeric(df_nulos['comienzo_atraccion'], errors='coerce')
df_nulos['comienzo_atraccion'] = df_nulos['comienzo_atraccion'].astype('Int64')
df_nulos.info()
```

```python
df_nulos.head(2)
```

```python
df_nulos.isnull().sum()
```

### Transformación 'comienzo_atraccion' a fechas concretas

```python
from datetime import datetime, timedelta
#Vamos a convertir el 'comienzo_atraccion' en un formato legible.
#Ponemos como fecha base el 1 de septiembre de 2022 a las 7:00h.
fecha_base = datetime (2022,9,1,7,0)
df_procedencia['comienzo_atraccion_fecha_hora'] = df_procedencia['comienzo_atraccion'].apply(lambda x: fecha_base + timedelta(hours=int(x)))
df_procedencia.head()
```

```python
#También para mayor legibilidad, colocamos la columna creada al lado de 'comienzo_atraccion'
df_procedencia = df_procedencia.iloc[:, [0, 1, 2, 5, 3, 4]]
df_procedencia.head()
```

## 3 - Descarga del documento limpio

```python
#Me descargo el csv con la columna de las fechas
df_procedencia.to_csv("id_visitante_atracciones_limpio.csv",index=False)
```

```python
df_sin_fechas = df_procedencia.drop(['comienzo_atraccion_fecha_hora'], axis=1)
df_sin_fechas.head()
```

```python
#Me descargo el csv sin la columna de las fechas
df_sin_fechas.to_csv("id_visitante_atracciones_limpio_sin_fechas.csv",index=False)
```

```python
#Me descargo el documento con nulos en limpio:
df_nulos.to_csv("id_visitante_atracciones_limpio_con_nulos.csv", index=False)
```
