## 1 - Importación y lectura del df

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/valoraciones.json')
```

```python
#Ejecuto el documento para ver cómo está dividido el df. Veo que las columnas son cada una de las fotografías y que en las filas tenemos información de cada una de ellas.
df.head()
```

```python
import json
with open('valoraciones.json', 'r', encoding="utf-8") as file:
    data = json.load(file)

type(data)
```

```python
data
```

```python
#Intento convertit el diccionario en dataframe, añadiendo los nombres de columnas:
df = pd.DataFrame(data, columns=['t_id','valoracion'])
df.head()
```

```python
#Como veo que no me coge los datos, voy a separar el diccionario en dos listas para luego juntarlas:
datos_1 = []
datos_2= []

for key, value in data.items():
  datos_1.append(key)
  datos_2.append(value)
```

```python
datos_1
```

```python
datos_2
```

```python
#Creo un df para la primera columna, añadiendo ya el nombre correspondiente:
df1 = pd.DataFrame(datos_1, columns=['t_id'])
df1.head()
```

```python
df2 = pd.DataFrame(datos_2, columns=['valoracion'])
df2.head()
```

```python
#Concatenamos ambos para poder tener un solo df:
df_concat = pd.concat([df1, df2], axis=1)
df_concat.head()
```
